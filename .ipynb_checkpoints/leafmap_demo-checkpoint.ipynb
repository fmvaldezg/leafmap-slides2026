# What leafmap Does Best
## A Quick Tour of leafmap's Unique Strengths

---

leafmap is not the right tool for every mapping task — but for certain use cases it has **no equal in the Python ecosystem**. This notebook demonstrates those strengths with minimal code.

| Strength | Why leafmap? |
|---|---|
| Cloud-Optimized GeoTIFFs (COG) | Stream satellite imagery directly from S3/cloud — no download needed |
| Split-panel comparison maps | Before/after or layer comparison with a draggable divider |
| WMS tile services | Load live layers from government and research servers |
| STAC catalog access | Browse and visualize spatiotemporal asset catalogs |
| Linked multi-panel maps | Synchronized views for comparing multiple datasets |

**Install:** `pip install leafmap`

---

```python
import leafmap
print(f"leafmap version: {leafmap.__version__}")
```

---
## 1. Cloud-Optimized GeoTIFFs (COG)

A **Cloud-Optimized GeoTIFF (COG)** is a satellite image stored on a cloud server in a way that lets you view any part of it without downloading the whole file. leafmap streams it directly into your notebook.

This example shows Maxar satellite imagery of Libya before and after the 2023 flooding disaster.

```python
# Load a single COG — satellite imagery of Libya after the 2023 flooding
# COG files stream from the cloud — no full download needed
url_after = "https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-09-13.tif"

m = leafmap.Map(center=[32.75, 22.63], zoom=13)
m.add_cog_layer(url_after, name="Libya — Post-flood (Sep 2023)")
m
```

---
## 2. Split-panel comparison map

leafmap's `split_map()` creates a side-by-side view with a **draggable divider**. This is one of the most powerful features for showing change over time — drag the slider to compare before and after.

Same location, two dates: Libya before and after the September 2023 flooding.

```python
# Before/after split map — drag the divider to compare
# The COG files cover the Derna region of Libya (32.75°N, 22.63°E)
# where catastrophic flooding occurred in September 2023
url_before = "https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-07-01.tif"
url_after  = "https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-09-13.tif"

m = leafmap.Map(
    center=[32.75, 22.63],
    zoom=13,
    height=500
)
m.split_map(
    left_layer=url_before,
    right_layer=url_after,
    left_label="Before — July 2023",
    right_label="After — September 2023 (flooding)"
)
m
```

```python
# Split map also works with basemaps — compare satellite vs street view
# Any basemap name from leafmap.basemaps works as left_layer or right_layer

m = leafmap.Map(center=[39.952, -75.165], zoom=13, height=500)
m.split_map(
    left_layer="Esri.WorldImagery",
    right_layer="CartoDB.Positron",
    left_label="Satellite",
    right_label="Streets"
)
m
```

---
## 3. WMS tile services

**WMS (Web Map Service)** is a standard protocol used by governments, research institutions, and NGOs to serve live map data over the web. leafmap can load any WMS layer in two lines.

This example loads the **NLCD (National Land Cover Database)** — a satellite-derived land cover map of the continental US — directly from the USGS server.

```python
# NLCD 2021 Land Cover — live from the USGS/MRLC WMS server
wms_url = "https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2021_Land_Cover_L48/wms?"

m = leafmap.Map(center=[40, -100], zoom=4)
m.add_wms_layer(
    wms_url,
    layers="NLCD_2021_Land_Cover_L48",
    name="NLCD 2021 Land Cover",
    format="image/png",
    transparent=True
)
# Add the built-in NLCD legend
m.add_legend(builtin_legend="NLCD", title="Land Cover Type")
m
```

---
## 4. Land cover change — split map with NLCD

leafmap has the full NLCD (National Land Cover Database) series pre-configured as built-in basemaps. Combine with `split_map()` to show how land cover changed across two decades — no WMS setup needed.

```python
# Compare NLCD 2001 vs 2021 land cover using built-in leafmap basemap names
# These NLCD layers are pre-configured in leafmap — no WMS setup needed

m = leafmap.Map(center=[36.1, -114.9], zoom=9, height=500)
m.split_map(
    left_layer="NLCD 2001 CONUS Land Cover",
    right_layer="NLCD 2021 CONUS Land Cover",
    left_label="2001",
    right_label="2021"
)
m.add_legend(builtin_legend="NLCD", title="Land Cover")
m
```

---
## 5. STAC — SpatioTemporal Asset Catalog

**STAC** is a standard for organizing large collections of satellite imagery. Major providers like NASA, Microsoft Planetary Computer, and AWS host petabytes of data in STAC catalogs. leafmap can load a STAC item directly from a URL.

This example loads multispectral SPOT satellite imagery of Canada.

```python
# STAC item — SPOT-5 orthoimage of northern Canada from 2007
stac_url = (
    "https://canada-spot-ortho.s3.amazonaws.com/canada_spot_orthoimages/"
    "canada_spot5_orthoimages/S5_2007/S5_11055_6057_20070622/"
    "S5_11055_6057_20070622.json"
)

# Check what bands are available and where the image is located
print("Available bands :", leafmap.stac_bands(stac_url))
```

```python
# SPOT-5 scene covers northern Alberta/Saskatchewan, Canada
# Center: approx 60.57°N, -110.55°W at zoom 10
m = leafmap.Map(center=[60.57, -110.55], zoom=10)

# True color (RGB) — loaded first, visible by default
m.add_stac_layer(stac_url, bands=["B3", "B2", "B1"], name="True Color (RGB)", vmin=0, vmax=150)

# False color — near-infrared highlights vegetation in red
# shown=False loads the layer but keeps it hidden until toggled on
m.add_stac_layer(stac_url, bands=["B4", "B3", "B2"], name="False Color (NIR)", vmin=0, vmax=150, shown=False)

# Layer control lets you toggle between True Color and False Color
m.add_layer_control()
m
```

---
## 6. Linked multi-panel maps

`leafmap.linked_maps()` creates multiple maps that are **synchronized** — when you pan or zoom one, all others follow. Useful for comparing datasets side by side.

```python
# Four NLCD time periods — all panning and zooming together
layers = [f"NLCD {year} CONUS Land Cover" for year in [2001, 2006, 2016, 2021]]
labels = [f"NLCD {year}" for year in [2001, 2006, 2016, 2021]]

leafmap.linked_maps(
    rows=2,
    cols=2,
    height="300px",
    layers=layers,
    labels=labels,
    center=[36.1, -115.2],
    zoom=9
)
```

---
## 7. DEM and raster visualization

leafmap can load local raster files (GeoTIFF) and visualize them with colormaps. This example loads a Digital Elevation Model (DEM) with a terrain colormap.

```python
# Download a sample DEM (SRTM 90m elevation)
dem_url = "https://github.com/opengeos/datasets/releases/download/raster/srtm90.tif"
leafmap.download_file(dem_url, "srtm90.tif", quiet=True)

m = leafmap.Map()
m.add_raster("srtm90.tif", colormap="terrain", layer_name="Elevation (m)")

# Generate colors from the terrain colormap for the colorbar
import matplotlib.pyplot as plt
import numpy as np
cmap = plt.get_cmap("terrain")
colors = ["#{:02x}{:02x}{:02x}".format(*[int(c*255) for c in cmap(v)[:3]]) for v in np.linspace(0, 1, 10)]

m.add_colorbar(colors=colors, vmin=0, vmax=4000, caption="Elevation (m)")
m
```

## Summary — When to use leafmap

| Use case | leafmap | Folium |
|---|---|---|
| Satellite imagery (COG/STAC) | ✅ Best choice | ❌ Not supported |
| Before/after split map | ✅ One line | ⚠️ Complex setup |
| WMS tile services | ✅ Simple | ✅ Works |
| Linked synchronized maps | ✅ Built-in | ❌ Not supported |
| Vector data (points, polygons) | ⚠️ Works in notebook, export issues | ✅ Best choice |
| Choropleth maps | ⚠️ Works but quirky | ✅ Best choice |
| HTML export for web | ⚠️ Widget-based, fragile | ✅ Clean Leaflet.js |

**Rule of thumb:** Use leafmap when you need satellite imagery, raster data, or split/linked map views inside a notebook. Use Folium when you need clean vector maps that export to reliable HTML.
